**Metodología CRISP – DM 2**
Pasos estándares en un máximo de 3 oraciones, por cada fase
1. Entendimiento del negocio – ¿Qué necesita el negocio? 
a. Se centra en la comprensión de los objetivos y requisitos del proyecto 
b.	Las tareas son: determine los objetivos comerciales, evaluar la situación, determine los objetivos de la minería de datos, producir plan de proyecto
c.	Establecer un sólido entendimiento empresarial es absolutamente esencial.
2.	Comprensión de datos: ¿qué datos tenemos/necesitamos? ¿Está limpio?
a.	Identificar, recopilar y analizar los conjuntos de datos que pueden ayudarlo a lograr los objetivos del proyecto
b.	Las tareas son: recopilar datos iniciales, describir los datos, explorar los datos y verificar la calidad de los datos
c.	Identificar los problemas de calidad de la data son claves para lograr la consistencia de la información.
3.	Preparación de datos: ¿cómo organizamos los datos para el modelado?
a.	Preparar los conjuntos de datos finales para el modelado.
b.	Las tareas son: seleccionar datos, limpiar datos, construir datos, integre datos y formatear datos.
c.	En esta fase obtendremos los datos limpios y con el formato correcto, así como suficientes para iniciar el modelado de datos.
4.	Modelado – ¿Qué técnicas de modelado debemos aplicar?
a.	construirá y evaluará varios modelos basados en varias técnicas de modelado diferentes
b.	Las tareas son: seleccionar técnicas de modelado, generar diseño de prueba, modelar compilación y evaluar el modelo.
c.	Se debe continuar iterando hasta encontrar un modelo "suficientemente bueno”
5.	Evaluación: ¿Qué modelo cumple mejor con los objetivos comerciales?
a.	Analiza de manera más amplia qué modelo se adapta mejor al negocio y qué hacer a continuación.
b.	Las tareas son: evaluar resultados, proceso de revisión, determine los próximos pasos.
c.	Es importante determinar si debe continuar con la implementación, iterar más o iniciar nuevos proyectos.
6.	Implementación: ¿cómo acceden las partes interesadas a los resultados?
a.	Un modelo no es particularmente útil a menos que el cliente pueda acceder a sus resultados
b.	Las tareas son: Planificar la implementación, Supervisión y mantenimiento del plan, Producir informe final y Revisar el proyecto
c.	A menudo se requiere un monitoreo constante y un ajuste ocasional del modelo.
